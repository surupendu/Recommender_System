```python
from surprise import Dataset,evaluate
from surprise import KNNBasic
```

```python
data = Dataset.load_builtin('ml-100k')
trainingSet = data.build_full_trainset()
```

```python
sim_options = {
    'name' : 'cosine',
    'user_based' : True
}

knn = KNNBasic(sim_options=sim_options)
```

```python
knn.train(trainingSet)
```

```python
testSet = trainingSet.build_anti_testset()
pred = knn.test(testSet)
```

```python
from collections import defaultdict

def top3_recommendations(pred,topN=3):
    recommend = defaultdict(list)
    for uid,iid,true_r,est,_ in pred:
        recommend[uid].append((iid,est))
    
    for uid,user_rating in recommend.items():
        user_rating.sort(key=lambda x:x[1],reverse=True)
        recommend[uid] = user_rating[:topN]
        
    return recommend
```

```python
import os,io

def read_item_names():
    file = open('C:\\Users\\DELL\\.surprise_data\\ml-100k\\ml-100k\\u.item','r',encoding='ISO-8859-1')
    movie_names = {}
    for line in file:
        line = line.split('|')
        movie_names[line[0]] = line[1]
        
    return movie_names
```

```python
recommend = top3_recommendations(pred)
```
